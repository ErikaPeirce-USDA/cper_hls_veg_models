```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iAPAR_fixed.csv')
```

```python
df['Pasture'] = df['Id'].apply(lambda x: x.split('_')[0])
```

```python
sns.lmplot(x='iAPAR_at_Sampling', y='Total_Biomass', hue='Treatment', col='Year', col_wrap=3,
          data=df.groupby(['Year', 'Treatment', 'Pasture']).mean().reset_index())
```
